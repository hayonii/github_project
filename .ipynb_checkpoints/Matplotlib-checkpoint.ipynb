## 스타일
- plot() 함수에 다양한 키워드 인수를 전달하여 스타일을 바꿀 수 있다.

```python
import numpy as np

plt.plot(np.arange(6), np.arange(6)**2,
        color='red',
        linestyle='-.', # solid(-), dashed(--), dotted(:), dashdot(-.)
        linewidth=2.5,
        marker='o',
        markeredgecolor='green',
        markerfacecolor='yellow',
        markersize=9.3)
plt.show()
```

- 상징 문자열을 이용해서 스타일을 적용할 수 있다.
- 주요 요소들에 대한 상징은 다음 표와 같다.      

```python
x = np.linspace(0, 2*np.pi, 10)

plt.plot(x, np.sin(x), 'ro--', x, np.cos(x), 'g^-.', x, -np.sin(x), 'bx:')
plt.xlim(0, 2*np.pi)
plt.show()
```

```python
x = np.arange(0.,5.,0.2)
plt.plot(x, x, 'r--', x, x**2, 'bs', x, x**3, 'g^:')
plt.show()
```

```python
x = np.arange(10)
y = np.random.randint(0, 100, 10) # 0에서 100 사이의 난수 정수 10개

plt.plot(x, y, c='b', lw=3, ls=':', marker='o', ms=10, mec='r', mfc='y', mew=2)
plt.title('Various Style')
plt.show()
```

## 제목 표시하기
- title() 함수를 이용해서 그림의 제목을 표시할 수 있다.
- xlabel(), ylabel() 함수를 이용해서 각각 x축의 이름과 y축의 이름을 표시할 수 있다.
- labelpad: 축 레이블의 여백을 지정할 때 사용하는 parameter, 단위는 pt

```python
plt.plot(['Jan', 'Feb', 'March'], [104.0, 77.5, 150.3], 'x-.')
plt.title('<Monthly Sales Graph>')
plt.xlabel('[Month]', labelpad=10)
plt.ylabel('[Sales]', labelpad=20)
plt.show()
```

## 범례 표시하기
- plot()함수에 label 키워드 인자를 설정함으로써 그래프에 이름을 지정할 수 있다.
- 이렇게 지정된 이름을 legend() 함수를 통해 범례로 표시할 수 있다.
- loc 파라미터에 다음 문자열을 지정하여 범례의 위치를 정할 수 있다.
<pre>
    'best' (최적의 위치에 알아서)
    'upper right'
    'upper left'
    'lower right'
    'lower left'
    'right'
    'center right'
    'center left'
    'lower center'
    'upper center'
    'center'
</pre>

```python
import numpy as np
import matplotlib.pyplot as plt

a = np.arange(0, 3, .02)
c = np.exp(a)
d = c[::-1]

plt.plot(a,c,'r--', label='Model length')
plt.plot(a,d,'b:', label='Data length')
plt.plot(a, c+d, 'g',label='Total message length')
plt.legend(loc='best', shadow=True )
plt.show()
```

## 서브플롯 그리기

### subplot() 함수
- subplot() 함수를 이용해서 하나의 그림에 여러 개의 그래프를 그릴 수 있다.
    - nrows, ncols, index의 인수를 이용하여 서브 플롯의 위치를 설정한다.

```python
x1 = np.linspace(0,5,50)
x2 = np.linspace(0,2,50)

y1 = np.cos(2*np.pi*x1)+np.exp(-1)
y2 = np.cos(2*np.pi*x2)

plt.subplot(1,2,1) #nrow=2, ncol=1, index=1, 2행 1열짜리 중 첫번째
plt.plot(x1, y1, 'o-')
plt.title('1st Graph')
plt.xlabel('time(s)')
plt.ylabel('Damped')

plt.subplot(1,22)
plt.plot(x2, y2, '.-')
plt.title('2nd Graph')
plt.xlabel('time(s)')
plt.ylabel('Undamped')

# plt.tight_layout() # 쇼서브플롯의 레이아웃을 자동으로 맞춰준다.
plt.show()
```

### subplots() 함수
- subplots() 함수는 Figure, Axes 객체를 반환한다.
- Axes 클래스 : 공간이 확보된 이미지 영역 정보를 가지고 있다.
    - 주요 인자
    > - nrows, ncols : 서브 플롯의 행과 열 크기 지정
    > - sharex, sharey : boolean 값(default : False), x, y축의 값 공유 여부 지정

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(1,4, sharey = True)

fig.set_size_inches((16,4))
axes[0].scatter(np.random.random(100), np.random.random(100), c = 'cyan', edgecolor = 'red', alpha = 0.5)
axes[1].scatter(np.random.random(100), np.random.random(100), c = 'magenta', edgecolor = 'green', alpha = 0.5)
axes[2].scatter(np.random.random(100), np.random.random(100), c = 'yellow', edgecolor = 'blue', alpha = 0.5)
axes[3].scatter(np.random.random(100), np.random.random(100), c = 'white', edgecolor = 'black', alpha = 0.5)

axes[0].set_title('subplor-0')
axes[1].set_title('subplor-1')
axes[2].set_title('subplor-2')
axes[3].set_title('subplor-3')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

fig, axes = plt.subplots(2,2, sharey = True)

fig.set_size_inches((8,8))
axes[0][0].scatter(np.random.random(100), np.random.random(100), c = 'cyan', edgecolor = 'red', alpha = 0.5)
axes[0][1].scatter(np.random.random(100), np.random.random(100), c = 'magenta', edgecolor = 'green', alpha = 0.5)
axes[1][0].scatter(np.random.random(100), np.random.random(100), c = 'yellow', edgecolor = 'blue', alpha = 0.5)
axes[1][1].scatter(np.random.random(100), np.random.random(100), c = 'white', edgecolor = 'black', alpha = 0.5)

axes[0][0].set_title('subplor-0')
axes[0][1].set_title('subplor-1')
axes[1][0].set_title('subplor-2')
axes[1][1].set_title('subplor-3')

plt.show()
```

# 막대 그래프 그리기

## 수직 막대 그래프

```python
x = np.arange(0,20)
y = x*2

x_ticks =  np.arange(len(x)) # x 축의 눈금을 표시할 index 값
plt.bar(x, y, label = 'Blue bar', color = 'purple', edgecolor = 'skyblue', linewidth = 3)
plt.xlabel('bar number')
plt.ylabel('bar height')
plt.title('Bar Chart Example')
plt.xticks(x_ticks, x_ticks+1)
plt.legend()
plt.show()
```

## 수평 막대 그래프

```python
x = np.arange(0,20)
y = x*2

y_ticks =  np.arange(len(x)) # y축의 눈금을 표시할 index 값

# 수평 막대 그래프를 그린다고 해서 x와 y의 값의 위치가 바뀌는 것은 아니다.
plt.barh(x, y, label = 'Blue bar', color = 'purple', edgecolor = 'skyblue', linewidth = 3)
plt.ylabel('bar number')
plt.xlabel('bar height')
plt.title('H-Bar Chart Example')
plt.yticks(y_ticks, y_ticks+1)
plt.legend(loc = 'lower right')
plt.show()
```

```python
label = ['Seoul', 'Busan', 'Incheon', 'Gwangju', 'Jeju']
men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x_ticks_index = np.arange(len(men_means))
width = 0.35 # 막대 그래프 폭의 크기
plt.bar(x_ticks_index-width/2, men_means, width = width, color = 'aquamarine', label = 'Men')
plt.bar(x_ticks_index+width/2, women_means, width = width, color = 'plum', label = 'Women')

plt.xlabel('City')
plt.ylabel('Score')
plt.title('Score by group and gender')
plt.xticks(x_ticks_index, label)
plt.legend()
plt.show()
```

# 산점도 그리기
- 산점도(scatter plot)는 직교좌표계를 이용해 두 변수 간의 상관관계를 확인하고자 할 때 쓰는 플롯이다.

```python
x1 = [2,3,4]
y1 = [5,5,5]

x2 = [1,2,3,4,5]
y2 = [2,3,2,3,5]

y3 = [6,8,7,8,7]

plt.scatter(x1, y1)
plt.scatter(x2, y2, marker = 'v', color = 'green')
plt.scatter(x2, y3, marker = '^', color = 'red')
plt.title('Scatter Plot Example')
plt.show()
```

```python
for color in ['blue', 'orange', 'green'] :
    n = 50
    x, y = np.random.rand(2,n) # [0,1) 사이의 난수 2 x 50 모양으로 생성]
    scale = 190.0 * np.random.rand(n) + 10 # [10, 200) 사이의 난수 50개 발생
    plt.scatter(x,y,c=color, s=scale, label=color, alpha=0.4, edgecolors='none')

plt.legend(loc = (1.04, 0.7))
plt.grid(True)
plt.show()
```

```python
# 히스토그램 그리기
- 자료의 분포를 몇 개의 구간으로 나눈 후 각 구간에 해당하는 데이터의 빈도수를 표로 나타낸 것을 도수분포표(frequency tavle)라고 한다.
- 히스토그램은 도수분포표에서 각 구간별 데이터의 빈도수를 막대그래프로 나타낸 것이다.
```

```python
y = np.random.randn(1000) # 평균이 0, 표준편차가 1인 표준정규분포 난수 1000개 생성
# print(np.mean(y), np.std(y))
print(np.max(y), np.min(y))
x = np.arange(len(y))
plt.bar(x,y)
plt.title('Raw Data <Bar Chart>')
plt.show()

plt.hist(y, bins = 50) # bins : 데이터를 분할할 구간 수
plt.title('Histogram')
plt.show()
plt. hist(y, cumulative=True, bins=20)
plt.title('Cumulative Histogram')
plt.show()
```

# 원 그래프 그리기

```python
import matplotlib.pyplot as plt

label = ['Pathom', 'Java', 'C']
# autopce : 각 조각 위에 비율(%)을 표시하는 형식 문자열을 지정한다.
plt.pie([0.3, 0.35, 0.4], labels=label, autopct='%.1f%%')
plt.title('Popular Language Ranking')
plt.legend(loc=(1, 0.3))
plt.show()
```

```python
data = [0.3, 0.35, 0.4]
label = ['Python', 'Java', 'C']
colors = ['coral', 'dodgerblue', 'yellowgreen']

plt.pie(
    data, # 조각의 크기를 결정하는 값
    labels = label, # 조각의 이름 표시
    explode = (0.1, 0, 0), # 조각을 중심으로부터 얼마나 떨어지게 할 것인지를 설정한다. (0.1 : 반지름의 10% 만틈 벗어남)
    colors = colors, # 조각의 색깔
    autopct = '%.2f%%', # 비율을 표시하는 형식 문자열
    shadow = True, # 그림자 표시 여부
    startangle = 90, # 조각의 시작 부분을 각도로 표시. 해당 각도에서부터 반시계 방향으로 회전하면서 조각을 표시(기본값 : 0, 3시 방향)
    counterclock=False, # 반시계방향으로 순서대로 표시할 지 결정 (기본값 : True)
    rotatelabels = True # 조각의 이름을 회전한다. (기본값 : False)
)
plt.title('Popular Language Ranking')
plt.show()
```

# 박스플롯 그리기
- 박스플롯은 데이터 집합의 범위와 중앙값 등 통계 수치 값을 확인하고자 하는 목적으로 사용한다.
- 이상치(outlier)가 있는지 검사하기 수월하다.
- 최소값 : 제 1사분위에서 1.5IQR(inter quartile range : 사분위수범위, Q3-Q1)을 뺀 위치
- 제 1사분위(Q1) : 중앙값 기준으로 하위 50% 중의 중앙값, 전체 데이터 중 하위 25%에 해당하는 값
- 제 2사분위(Q2) : 50%의 위치에 중앙값(median)을 의미
- 제 3사분위(Q3) : 중앙값 기준으로 상위 50% 중의 중앙값, 전체 데이터 중 상위 25%에 해당하는 값
- 최대값 : 제3사분위에서 1.5IQR을 더한 위치
- 최대값, 최소값 범위를 벗어나는 값을 이상치 데이터로 간주한다.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
data_a = np.random.normal(0, 2.0, 1000) # 평균 : 0 , 표준편차 : 2 , 1000개)
data_b = np.random.normal(-3.0, 1.5, 500) # 평균 :-3, 표준편차 : 1.5, 500개
data_c = np.random.normal(1.2, 1.5, 1500)

plt.boxplot([data_a, data_b, data_c])
plt.ylim(-10.0, 10.0)
plt.xlabel('Data Type')
plt.ylabel('Value')
plt.show()
```

```python
data_a = np.random.normal(0, 2.0, 1000)
print(f'min: {np.min(data_a):.2f}, median:{np.median(data_a):.2f}, max : {np.max(data_a):.2f}')
quantile = np.quantile(data_a, [0, 0.25, 0.5, 0.75, 1])
print(quantile)
IQR = quantile[3] - quantile[1]
print('IQR:', IQR)
```
